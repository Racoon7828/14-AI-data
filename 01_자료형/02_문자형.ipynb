```python
#문자열 포매팅 (비주류)
#1. 숫자 바로 대입
print("I eat %d apples" %3)

#2. 문자열 바로 대입
print("I eat %s apples" % "five")

#3. 숫자 값을 나타내는 변수로 대입
number = 3
print("I eat %s apples" % number)

#4. 두개 이상의 값 넣기
number = 10
day = "three"
print("I ate %d apples. so I was sick for %s days." % (number, day))
```

### 문자열 포맷코드 ###
- %s : 문자열(String)
- %c : 문자 1개(character)
- %d : 정수(Integer)
- %f : 실수(floating-point)
- %o : 8진수
- %x : 16진수
- %% : Literal % (문자 % 자체)

```python
# format 함수를 사용한 포매팅 (비주류)
# 1. 숫자 바로 대입하기
print("I eat {0} apples".format(3))

# 2. 문자열 바로 대입하기
print("I eat {0} apples".format("five"))

# 3. 숫자 값을 가진 변수로 대입하기
number = 3
print("I eat {0} apples".format(number))

#4. 두개 이상의 값 넣기
number = 10
day = "three"
print("I ate {0} apples. so I was sick for {1} days.".format(number, day))
```

```python
# f문자열 포매팅 (파이썬 3.6 이상) (주류)
name = '홍길동'
age = 30
print(f'나의 이름은 {name}입니다. 나이는 {age}입니다.')
print(f'나는 내년이면 {age + 1}살이 된다.')
```

```python
# 문자열 관련 함수들

# 1. 문자 개수 세기 - count
a = "hobby"
print(a.count("b"))

# 2. 위치 알려주기 - find
b = "Python is the best choice"
print(b.find("b"))
print(b.find("k"))

# 3. 위치 알려주기 = index
print(b.index("b"))
# 문자열 안에 존재하지 않는 문자를 찾으면 오류 발생

# 4. 문자열 삽입 - join
print(",".join('abcd'))

# 5. 소문자를 대문자로 바꾸기 - upper
print(a.upper())

# 6. 대문자를 소문자로 바꾸기 - lower
a = 'HOBBY'
print(a.lower())

# 7. 공백 지우기 - strip
a = " hi "
print(a.strip())
# 8. 왼쪽 공백 지우기 - lstrip
# 9. 오른쪽 공백 지우기 - rstrip

# 10. 문자열 바꾸기 - replace
a = "life is too short"
print(a.replace("life", "Your leg"))

# 11. 문자열 나누기 - split
print(a.split())
```
