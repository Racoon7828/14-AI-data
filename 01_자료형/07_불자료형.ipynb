```python
# 불(bool) 자료형이란 참(True)과 거짓(False)을 나타내는 자료형이다.
```

```python
# 불 자료형은 어떻게 사용할까?
a = True
b = False
print(type(a))
print(a)
```

```python
print(1 == 1)
print(2 > 1)
print(2 < 1)
```

```python
print(1, bool(1))
print(0, bool(0))
print(2, bool(2))
print("python", bool("python"))
print("", bool(""))
print("[]", bool([]))
print("[1,2,3]", bool([1,2,3]))
```

```python
if [1,2,3]:
	print("True")
else:
	print("False")

if 3 < 1:
	print("True")
else:
	print("False")
```

```python
# 논리 연산자
# 01 and 연산자
print(True and True)
print(True and False)
print(False and True)
print(False and False,'\n')

# 02 or 연산자
print(True or True)
print(True or False)
print(False or True)
print(False or False,'\n')

# 03 not 연산자
print(not True)
print(not False)
print(not 1)
print(not 0)
```
