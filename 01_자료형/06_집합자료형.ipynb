```python
# 집합 자료형은 어떻게 만들까? - set
a = set([1,2,3])
print(a)

# set()의 괄호 안에 리스트를 입력하거나, 다음과 같이 문자열을 입력해서 만들 수도 있다.
b = set('hello')
print(b)
# 중복 허용 X

# 중괄호({})를 사용해서 집합 자료형을 직접 만들 수도 있다.
c = {1,2,3}
print(type(c))
print(c)

# 순서가 없다
d = {'b','a','c'}
print(d)
```

```python
# 교집합
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])
print(s1.intersection(s2)) # intersection
print(s1 & s2) # &
```

```python
# 합집합
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])
print(s1.union(s2)) # union
print(s1 | s2) # | -> or 연산자?
```

```python
# 차집합
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])
print(s1.difference(s2)) # difference
print(s1 - s2) # -
```

```python
# 집합 관련 함수
# 01. 값 1개 추가하기 - add
s1 = set([1,2,3])
s1.add(4)
print(s1)

# 02. 값 여러 개 추가하기 - update
s1 = set([1,2,3])
s1.update([4,5,6])
print(s1)

# 03. 특정 값 제거하기 - remove
s1 = set([1,2,3])
s1.remove(2)
print(s1)

# 04. 모든 값 제거하기 - clear
s1 = set([1,2,3])
s1.clear()
print(s1)
```
