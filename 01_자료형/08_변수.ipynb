```python
# 변수를 만드는 여러가지 방법
a, b = ["python", "like"]
print(a)
print(b)
print(type(a))
print(type(b))
```

```python
(a, b) = "python", "like"
print(a)
print(b)
print(type(a))
print(type(b))
```

```python
a = b = "python"
print(a)
print(b)
print(type(a))
```
