```python
# 파이썬 함수의 구조
# def 함수_이름(매개변수):
# 수행할_문장1
# 수행할_문장2
# ...
```

```python
# 매개변수와 인수
# 매개변수는 함수에 입력으로 전달된 값을 받는 변수
# 인수는 함수를 호출할 때 전달하는 입력값

def add(a, b): # a, b는 매개변수
    return a + b

print(add(3, 4)) # 3, 4는 인수
```

```python
# 일반적인 함수
# 입력값이 있고 반환값이 있다.

def add(a, b):
    result = a + b
    return result
sum = add(3, 4)
print(sum)
```

```python
# 입력값이 없는 함수
# 입력값 X 반환값 O
def say():
    return 'Hi'
a = say()
print(a)
```

```python
# 반환값이 없는 함수
# 입력값 O 반환값 X
def add(a, b):
    print(f'{a},{b}의 합은 {a+b}입니다')
sum = add(3, 4)
print(sum)
```

```python
# 입력값도, 반환값도 없는 함수
# 입력값 X 반환값 X
def say():
    print('Hi')
say()
a = say()
print(a)
```

```python
# 매개변수를 지정하여 호출하기
def sub(a, b):
    return a - b
result = sub(a = 7, b = 3)
print(result)

result2 = sub(b = 7, a=3)
print(result2)

print("HI", "Bye", "Hello", sep="---")
```

```python
# 입력값이 몇 개가 될지 모를때는 어떻게 할까?
def add_many(*args):
    result = 0
    print(type(args))
    print(args)
    for i in args:
        result += i
    return result

result = add_many(1,2,3,4,5)
print(result)
```

```python
def add_mul(choice, *args):
    if choice == "add":
        result = 0
        for i in args:
            result += i
    elif choice == "mul":
        result = 1
        for i in args:
            result *= i
    return result

result = add_mul("add", 1,2,3,4,5)
print(result)

result = add_mul("mul", 1,2,3,4,5)
print(result)
```

```python
# 키워드 매개변수 - kwargs
# '키워드=값' 형태로 전달하는 매개변수를 받을 때 사용한다.
# 키워드 매개변수를 사용할 때는 매개변수 앞에 별 2개(**)를 붙인다.

def print_kwargs(**kwargs):
    print(kwargs)

print_kwargs(a=1)

print_kwargs(name='foo', age=20, job="student")
```

```python
def create_profile(**kwarg):
    print("=== 프로필정보 ===")
    for key, value in kwarg.items():
        print(f'{key}: {value}')

create_profile(이름='김철수', 나이=30, 직업='프로그래머', 취미='축구')
```

```python
# 함수의 반환값은 언제나 하나이다
def add_and_mul(a,b):
    return a+b, a*b

result = add_and_mul(3,4)
print(result)
```

```python
def add_and_mul(a,b):
    return a+b
    return a*b
result = add_and_mul(3,4)
print(result)
```

```python
def say_nick(nick):
    if nick == "학생":
        return
    print(f'너의 별명은 {nick}이다.')

say_nick('선생')
say_nick('학생')
say_nick('학부모')
```

```python
# 매개변수의 초깃값 미리 설정하기

def say_myself(name, age, man=True):
    print(f'너의 이름은 {name}이다.')
    print(f'너의 나이는 {age}살이다.')
    if man:
        print('남자입니다.')
    else:
        print('여자입니다.')

say_myself('홍길동', 25)
say_myself('김영희', 27, False)
```

```python
# 함수 안에서 선언한 변수의 효력 범위
a = 1
def vartest(a):
    a += 1 # 지역변수

vartest(a)
print(a)
```

```python
# 함수 안에서 함수 밖의 변수를 사용하는 방법
# 잘안씀
a = 1
def vartest():
    global a 
    a += 1
vartest()
print(a)
```

```python
# lambda 예약어
# lambda는 함수를 생성할 때 사용하는 예약어로, def와 동일한 역할을 한다.
# 보통 함수를 한 줄로 간결하게 만들 때 사용한다.
# def를 사용해야 할 정도로 복잡하지 않거나 def를 사용할 수 없는 곳에 주로 쓰인다.

# 함수_이름 = lambda 매개변수1, 매개변수2, ... : 매개변수를_이용한_표현식

def add(a,b):
    return a+b
result1 = add(3,4)
print(result1)

add = lambda a, b: a+b
result2 = add(3,4)
print(result2)
```

```python
# 함수의 독스트링(Docstring)
def add(a, b):
    """
    두 숫자를 더하는 함수

    Parameters:
    a (int, float): 첫 번째 숫자
    b (int, float): 두 번째 숫자

    Returns:
    int, float: 두 숫자의 합
    """
    return a + b

# 독스트링 확인하기
print(add.__doc__)
```
