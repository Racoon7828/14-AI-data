```python
numbers = [64, 34, 25, 12, 22, 11, 90]
words = ["banana", "Apple", "cherry", "date"]

numbers.sort(); print(numbers)
numbers.sort(reverse=True); print(numbers)
words.sort(); print(words) # 대문자가 소문자보다 우선합니다.
words.reverse(); print(words)
```

```python
a = (1, 2, 3, 4, 5)
b = ("red", "green", "blue")
c = (10,)  # 요소가 하나일 땐 쉼표 필수!
d1 = ()
d2 = tuple()
e = 100, 200, 300 # 괄호 없는 튜플 생성(Packing)

tuples = [a,b,c,d1,d2,e]
for t in tuples:
    print(f"길이: {len(t)}, 자료형: {type(t)}")
```
