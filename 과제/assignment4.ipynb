```python
### 문제 2-1: 학점 계산기

# 점수를 입력받아 다음 기준으로 학점을 출력하세요:

# - 90점 이상: A
# - 80점 이상 90점 미만: B
# - 70점 이상 80점 미만: C
# - 60점 이상 70점 미만: D
# - 60점 미만: F
score = int(input("점수를 입력하세요: "))
#   **예상 출력:**
#   점수를 입력하세요: 85
#   학점: B

# 여기에 코드를 작성하세요
if score >= 90:
    print("학점: A")
elif score >= 80:
    print("학점: B")
elif score >= 70:
    print("학점: C")
elif score >= 60:
    print("학점: D")
else:
    print("학점: F")
```

```python
### 문제 7-1: 택시 요금 계산기

# 거리(km)와 시간대를 입력받아 요금을 계산하세요.

# - 기본요금: 3000원 (2km까지)
# - 추가요금: 2km 초과시 1km당 1000원
# - 심야할증(22시~04시): 총 요금의 20% 추가
distance = float(input("거리를 입력하세요 (km): "))
hour = int(input("현재 시간을 입력하세요 (0-23): "))
#   **예상 출력:**
#   거리를 입력하세요 (km): 5
#   현재 시간을 입력하세요 (0-23): 23
#   요금: 7200원

# 여기에 코드를 작성하세요
base_fare = 3000
if distance > 2:
    fare = base_fare + (distance - 2) * 1000
else:
    fare = base_fare
if 22 <= hour or hour < 4:
    fare *= 1.2
print(f"요금: {int(fare)}원")
```

```python
### 윤년 판별기

# 연도를 입력받아 윤년인지 아닌지 판별하세요.

# 윤년 조건:
# 1. 4로 나누어떨어지는 해는 윤년
# 2. 단, 100으로 나누어떨어지는 해는 평년
# 3. 단, 400으로 나누어떨어지는 해는 윤년

year = int(input("연도를 입력하세요: "))

# **예시:**
# - 2020년: 윤년 (4로 나누어떨어짐)
# - 2100년: 평년 (100으로 나누어떨어지지만 400으로는 안됨)
# - 2000년: 윤년 (400으로 나누어떨어짐)

if year % 400 == 0:
    print("윤년")
elif year % 100 == 0:
    print("평년")
elif year % 4 == 0:
    print("윤년")
else:
    print("평년")
```
