```python
# 1) 정수 1, 2, 3, 4, 5를 담은 튜플을 생성하시오.
a = (1,2,3,4,5)

# 2) 문자열 "red", "green", "blue"를 담은 튜플을 생성하시오.
b = ("red", "green", "blue")

# 3) 요소가 하나뿐인 튜플을 올바르게 생성하시오. (주의: 쉼표 필요)
c = (42,)

# 4) 빈 튜플을 두 가지 방법((), tuple())으로 생성하시오.
d = ()
e = tuple()

# 5) 괄호 없이 튜플을 생성하시오. (패킹)
f = 1,2,3

# 6) 각 튜플의 길이(len)와 자료형(type)을 출력하시오.
print("a:", "길이:", len(a), "자료형:", type(a))
print("b:", "길이:", len(b), "자료형:", type(b))
print("c:", "길이:", len(c), "자료형:", type(c))
print("d:", "길이:", len(d), "자료형:", type(d))
print("e:", "길이:", len(e), "자료형:", type(e))
print("f:", "길이:", len(f), "자료형:", type(f)) 
```

```python
# 1. 다음 정보를 담는 딕셔너리를 만드시오
#    - 이름: 본인 이름
#    - 나이: 본인 나이
#    - 취미: 본인 취미 (리스트로 여러 개)
#    - 거주지: 본인 거주지
profile = {
    '이름': '김병욱',
    '나이': 27,
    '취미': '게임',
    '거주지': '대구'
}
print(profile)

# 2. 위 딕셔너리에 '전공' 항목을 추가하시오
profile['전공'] = '컴퓨터공학'

# 3. '취미' 항목에 새로운 취미를 추가하시오
profile['취미'] = [profile['취미'], '유튜브 시청']

# 4. '나이' 항목의 값을 1 증가시키시오
profile['나이'] += 1

# 5. 모든 key와 value를 출력하시오
print(profile.keys())
print(profile.values())
```

```python
fruits = {
    'apple': 1000,
    'banana': 1500,
    'orange': 2000,
    'grape': 3000
}
print(fruits)

# 1. 모든 과일 이름(key)을 리스트로 출력하시오
print(list(fruits.keys()))

# 2. 모든 가격(value)을 리스트로 출력하시오
print(list(fruits.values()))

# 3. 'banana'의 가격을 출력하시오
print(fruits['banana'])

# 4. 'melon'이 딕셔너리에 있는지 확인하시오
print('melon' in fruits)

# 5. 'grape'를 pop() 함수를 사용하여 삭제하고, 삭제된 값을 출력하시오
temp = fruits.pop('grape')
print(temp)
```

```python
text = "python is great python is fun python is powerful"

# 요구사항:
# 1. 문자열을 공백으로 분리하여 단어 리스트 만들기
# 2. 각 단어의 등장 횟수 세기
# 3. 결과를 출력

# **출력 예시:**
# python: 3번
# is: 3번
# great: 1번
# fun: 1번
# powerful: 1번

text_list = text.split()
print(text_list)
print(f"python: {text_list.count('python')}번")
print(f"is: {text_list.count('is')}번")
print(f"great: {text_list.count('great')}번")
print(f"fun: {text_list.count('fun')}번")
print(f"powerful: {text_list.count('powerful')}번")
```

```python
a_class = {'김철수', '이영희', '박민수', '최지영', '홍길동'}
b_class = {'박민수', '최지영', '정수민', '강호동', '이영희'}

# 1. 두 반 모두에 있는 학생 (교집합)
print(a_class&b_class)

# 2. 두 반을 합친 전체 학생 (합집합)
print(a_class|b_class)

# 3. A반에만 있는 학생 (차집합)
print(a_class-b_class)

# 4. B반에만 있는 학생 (차집합)
print(b_class-a_class)
```
