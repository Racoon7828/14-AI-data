```python
# 리스트는 [], 튜플은 ()으로 둘러싼다.
# 리스트는 요솟값의 생성, 삭제, 수정이 가능하지만, 튜플은 요솟값을 바꿀 수 없다.
# 프로그램 실행 중 요솟값이 변하지 않기를 바란다면 튜플을 사용하자.
# 반대로 값을 수시로 변경해야 한다면 리스트를 사용하자
```

```python
#튜플 요솟값을 삭제하려 할 때
t1 = (1,2,'a','b')
del t1[0]
# 튜플은 요솟값을 지울 수 없다.
```

```python
# 투플 요소값을 변경하려 할 때
t1 = (1,2,'a','b')
t1[0] = 'c'
# 튜플은 요솟값을 바꿀 수 없다.
```

```python
a = 10
print(id(a))
b = 20
print(id(b))
```

```python
a = [1,2,3]
print(id(a))
	 
a.append(4)
print(a)
print(id(a))
```

```python
#튜플에 인덱싱 하기
a = (1,2,3,4)
print(a[0])
```

```python
# 튜플 슬라이싱하기
a = (1,2,3,4)
print(a[2:])
```

```python
# 튜플 더하기
a = (1,2)
b = (3,4)
c = a + b
print(c)
```

```python
# 튜플 곱하기
a = (3,4)
b = a * 3
print(b)
```

```python
# 튜플 길이 구하기
a = (1,2,3,4)
print(len(a))
```

```python
a = (1,2,3,4)
a.append(5, 6)
print(a)
```

```python
a = (1, 2, 3)
a.insert(0, 4)
print(a)
```
