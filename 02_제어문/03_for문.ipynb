```python
# for 문의 기본 구조
# for 변수 in 리스트(또는 튜플, 문자열):
#     수행할_문장1
#     수행할_문장2

num_list = ['one', 'two', 'three']
for num in num_list:
    print(num)
```

```python
a = [(1,2), (3,4), (5,6)]
for i in a:
    print(i)
```

```python
a = [(1,2), (3,4), (5,6)]
for first, second in a:
    print(first, second)
```

```python
marks = [90, 25, 67, 45, 80]
number = 0

for mark in marks:
    number += 1
    if mark >= 60:
        print(f'{number}학생은 통과입니다')
    else:
        print(f'{number}학생은 불통과입니다')
```

```python
# for문과 continue문
marks = [90, 25, 67, 45, 80]
number = 0

for mark in marks:
    number += 1
    if mark < 60:
        continue
    else:
        print(f'{number}학생은 통과입니다')
```

```python
result = 0
for number in range(1, 10+1):
    result += number
print(result)
```

```python
marks = [90, 25, 67, 45, 80]

for number in range(len(marks)):
    if marks[number] < 60:
        continue
    print(f'{number+1}학생은 통과입니다')
```

```python
# for와 range를 이용한 구구단
for i in range(2, 10):
    for j in range(1, 10):
        print(i * j, end=' ')
    print('')
```

```python
for i in range(2, 10):
    for j in range(1, 10):
        result = i * j
        print(f'{i} * {j} = {result}')
    print('')
```

```python
for i in range(1, 6):
    print("*"*i)
```

```python
for i in range(1, 6):
    print(" "* (5 - i), "*"*i)
```

```python
# 리스트 컴프리헨션 사용하기
a = [1,2,3,4]
result = []
for num in a:
    result.append(num * 3)
print(result)
```

```python
a = [1,2,3,4]

result = [num * 3 for num in a]
print(result)
```

```python
# for문과 break문
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# enumerate()함수 활용
fruits = ['apple', 'banana', 'orage']
for i, fruit in enumerate(fruits):
    print(f'{i+1}번째 : {fruit}')
```

```python
# zip 함수로 여러 리스트 함께 순회하기
names = ['홍길동', '유재석', '강호동']
scores = [60, 70, 80]
for name, score in zip(names, scores):
    print(f'{name}, {score}')
```

```python
# 3개 이상의 리스트 - zip()
names = ['홍길동', '유재석', '강호동']
Korean = [60, 70, 80]
English = [80, 90, 70]

for name, score1, score2 in zip(names, Korean, English):
    print(f'{name}: 국어:{score1} 영어:{score2}')
```
