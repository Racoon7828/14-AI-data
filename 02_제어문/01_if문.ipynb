```python
# 만약 주머니에 돈이 있으면 택시를 타고 가고, 없으면 걸어가라.
pocket = ['paper', 'cellphone', 'money']

if 'card' in pocket:
    print("택시를 타고 간다")
else:
    print("걸어간다")
```

```python
# 다양한 조건을 판단하는 elif
# 주머니에 돈이 있으면 택시를 타고 가고
# 주머니에 돈이 없지만 카드가 있으면 택시 타고
# 둘다 없으면 걸어가기

pocket = ['paper', 'cellphone', 'money']
card = True

if 'money' in pocket:
    print("택시를 타고 간다\n")

else:
    if card:
        print("택시\n")
    else:
        print("걸어간다\n")

if 'money' in pocket:
    print("택시를 타고 간다")

elif card:
    print("택시를 타고 간다")

else:
    print("걸어간다")
```

```python
# 성적이 90점 이상 'A학점'
# 성적이 90점 미만 80점 이상 'B학점'
# 성적이 80점 미만 70점 이상 'C학점'
# 성적이 70점 미만 60점 이상 'D학점'
# 그외 'F학점'

score = 60

if score >= 90:
    print('A학점')
elif 90 > score >= 80:
    print('B학점')
elif 80 > score >= 70:
    print('C학점')
elif score >= 60:
    print('D학점')
else:
    print('F학점')  
```

```python
# match-case 구문

# match 변수
#   case 패턴1:
#       수행할 문장
#   case 패턴2:
#       수행할 문장
```

```python
grade = input("A~F 중 하나의 값을 입력하시오")

match grade:
    case "A" | "B" | "C":
        print('합격입니다.')
    case _:
        print('불합격입니다.')
```
