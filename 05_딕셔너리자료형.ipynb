```python
# 딕셔너리는 Key와 Value를 한 쌍으로 가지는 자료형이다.
# {키:값} 형태로 데이터를 저장한다.
```

```python
# 딕셔너리는 어떻게 만들까?
dic = {'name':'kim', 'phone':'01012341234', 'birth':'0101'}
print(dic)
print(type(dic))
```

```python
# 딕셔너리 쌍을 추가하기
a = {1:'a'}
a[2]='b'
print(a)
a['name'] = 'ken'
print(a)
a[1]='c'
print(a)
a[(1,2)] = 'hobby'
print(a)
a[[3,4]] = 'job'
print(a)
```

```python
# 딕셔너리 쌍을 삭제하기
print(a)
del a[1]
print(a)

del a['name']
print(a)
```

```python
# 딕셔너리에서 Key를 사용해 Value 얻기
profile = {'name':'홍길동','age': 25}

profile['name']
```

```python
# 딕셔너리 관련 함수 /중요/
# 1. Key 리스트 만들기 - keys
profile = {'이름':'홍길동','나이': 25,'번호':'010-1234-1234'}
print(profile.keys())

# 2. Value 리스트 만들기 - values
profile = {'이름':'홍길동','나이': 25,'번호':'010-1234-1234','생일':'0101'}
print(profile.values())

# 3. Key, Value 쌍얻기 - items
profile = {'이름':'홍길동','나이': 25,'번호':'010-1234-1234'}
print(profile.items())

# 4. Key: Value 쌍 모두 지우기 - clear
profile = {'이름':'홍길동','나이': 25,'번호':'010-1234-1234'}
print(profile.clear())

# 5. Key로 Value 얻기 - get = 딕셔너리에서 Key를 사용해 Value 얻기
profile = {'이름':'홍길동','나이': 25,'번호':'010-1234-1234'}
print(profile.get('이름'))

# 6. 해당 Key가 딕셔너리 안에 있는지 조사하기 - in
profile = {'이름':'홍길동','나이': 25,'번호':'010-1234-1234'}
print('이름' in profile)

# 7. Key로 Value 얻기 - pop
profile = {'이름':'홍길동','나이': 25,'번호':'010-1234-1234'}
temp = 이름 = profile.pop('이름')
print(temp)
print(profile)
```
