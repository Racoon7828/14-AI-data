```python
# 리스트는 어떻게 만들까?
# 형식

a = []
b = [1, 2, 3]
c = ['Life', 'is', 'too', 'short']
d = [1, 2, 'Life', 'is']
e = [1, 2, ['Life', 'is']]
print(a)
print(len(a))
print(len(b))
print(len(c))
print(len(d))
print(len(e))
print(e[2])
```

```python
# 리스트 인덱싱
a = ['Life', 'is', 'too', 'short']
print(a[2])
print(a[-1])
print(a[0][3])

```

```python
a = [1,2,3]
print(a[0]+a[2])
```

```python
a = [1,2,3,[4,5,6]]
print(a[-1])
```

```python
# 리스트 슬라이싱
a = [1,2,3,4,5]
a[0:1+1]
```

```python
a[-2:1]
```

```python
# 리스트 연산
# 1. 더하기
a = [1,2,3]
b = [4,5,6]
print(a + b)

# 2. 곱하기(반복)
a = [1,2,3]
print(a*3)

# 3. 길이
len(a)
```

```python
# 리스트의 수정과 삭제
# 1. 값 수정하기
a = [1,2,3]
a[2] = 4
print(a)

# 2. 삭제하기
a = [1,2,3]
del a[1]
print(a)

a = [1,2,3,4,5]
del a[2:] # 슬라이싱
print(a)
```

```python
# 리스트 관련 함수
# 1. 리스트에 요소 추가하기 - append
a = [1,2,3]
a.append(4)
print(a)
a.append([5,6])
print(a)

# 2. 리스트 정렬 - sort
a = [1,4,3,2]
a.sort()
print(a)

# 3. 리스트 뒤집기 - reverse
a = ['a', 'c', 'b']
a.reverse()
print(a)

# 4. 인덱스 반환 - index
a = [1,2,3]
print(a.index(3))
print(a.index(2))

# 5. 리스트에 요소 삽입 - insert
a.insert(0,4)
print(a)

a = [1,2,3,[4,5,6]]
a[3].insert(-1,4)
print(a)

# 6. 리스트 요소 제거 - remove
a = [1,2,3,1,2,3]
a.remove(3)
print(a)

# 7. 리스트 요소 꺼내기 - pop
a = [1,2,3]
a.pop()
print(a)

# 8. 리스트에 포함된 요소 x의 개수 세기 - count
a = [1,2,3,1]
a.count(1)
print(a)

# 9. 리스트 확장 = extend
a = [1,2,3]
a.extend([4,5])
print(a)
b = [6,7]
a.extend(b)
print(a)
```
